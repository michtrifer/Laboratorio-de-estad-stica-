2.   Usar un decision tree y ajustar con dataset (r^2)
3.   Comparar R^2. Escribe observaciones
4.   Repetir (pasos del 1-3) con train test split (random_state=42) (R^2 train, test)
5.   Repetir (pasos del 1-3) con polynomialfeatures(degree=2)

```python
from sklearn.tree import DecisionTreeRegressor 
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
X = df.drop(columns=["sales"])
y = df["sales"]
```

### 1.

```python
lr=LinearRegression()
lr.fit(X,y)

y_pred = lr.predict(X)

print("R2:", r2_score(y, y_pred))
```

### 2.

```python
tree = DecisionTreeRegressor(random_state=42)

tree.fit(X, y)

y_pred_tree = tree.predict(X)

r2_tree = r2_score(y, y_pred_tree)
print("R2 (Decision Tree):", r2_tree)
```

### 3.

La regresión lineal explica el 89.7% de las ventas, mostrando buen ajuste y capacidad de generalización. El árbol de decisión tiene un R2 de 1, lo que indica un ajuste perfecto, pero probablemente refleja sobreajuste, por lo que podría no generalizar bien a nuevos datos.

### 4.

```python
Xtrain, Xtest, ytrain, ytest=train_test_split(X,y,test_size=.3,random_state=137)
```

```python
lr=LinearRegression()
lr.fit(Xtrain,ytrain)

y_pred_train = lr.predict(Xtrain)
y_pred_test = lr.predict(Xtest)

print("R2 train:", r2_score(ytrain, y_pred_train))
print("R2 test:",  r2_score(ytest, y_pred_test))
```

```python
tree = DecisionTreeRegressor(random_state=42)

tree.fit(Xtrain, ytrain)

y_pred_train = tree.predict(Xtrain)
y_pred_test = tree.predict(Xtest)

print("R2 train:", r2_score(ytrain, y_pred_train))
print("R2 test:",  r2_score(ytest, y_pred_test))
```

El modelo de regresión lineal logra un buen ajuste, explicando alrededor del 90% de los datos de entrenamiento y 87.8% de los de prueba. Esto indica que generaliza bien y mantiene un rendimiento estable.

El árbol de decisión tiene un R2 muy similar, aunque la regresión lineal muestra un equilibrio más consistente entre ajuste y generalización.

### 5.

```python
poly = PolynomialFeatures(degree=2)

Xtrain_poly = poly.fit_transform(Xtrain)
Xtest_poly = poly.transform(Xtest)

lr = LinearRegression()
lr.fit(Xtrain_poly, ytrain)

y_pred_train = lr.predict(Xtrain_poly)
y_pred_test = lr.predict(Xtest_poly)

print("R2 train:", r2_score(ytrain, y_pred_train))
print("R2 test:",  r2_score(ytest, y_pred_test))
```

```python
poly = PolynomialFeatures(degree=2)

Xtrain_poly = poly.fit_transform(Xtrain)
Xtest_poly = poly.transform(Xtest)

tree = DecisionTreeRegressor(random_state=42)
tree.fit(Xtrain_poly, ytrain)

y_pred_train = tree.predict(Xtrain_poly)
y_pred_test = tree.predict(Xtest_poly)

print("R2 train:", r2_score(ytrain, y_pred_train))
print("R2 test:",  r2_score(ytest, y_pred_test))
```

La regresión polinómica logra un muy buen ajuste y generaliza bien (R2 train: 0.99, test: 0.98).
El árbol con polinomios tiene un ajuste perfecto en entrenamiento (R2=1.0) y un test ligeramente mejor, pero muestra signos de sobreajuste.
