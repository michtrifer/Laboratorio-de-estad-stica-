```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
df = pd.read_excel("Boston Housing Dataset 1978.xlsx")
matriz = df.to_numpy()
```

### Limpieza y quitar duplicados 

```python
data=df.dropna()
data=data.drop_duplicates()
```

```python
data
```

### División de los datos 

```python
n = len(data)

mitad1 = np.random.choice(n, size=n//2, replace=False)

mitad2 = np.setdiff1d(np.arange(n), mitad1)
```

```python
data1 = data.iloc[mitad1]
data2 = data.iloc[mitad2]
matriz1 = data1.to_numpy()
matriz2 = data2.to_numpy()
```

```python
ytrain = matriz1[:, 13].reshape(-1, 1)
Xtrain = matriz1[:, :-1]
ytest = matriz2[:, 13].reshape(-1, 1)
Xtest = matriz2[:, :-1]
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler= StandardScaler().fit(Xtrain)
Xscaled=scaler.transform(Xtrain)
```

```python
columna_unos = np.ones((matriz1.shape[0], 1), dtype=int)
Xscaled = np.hstack([columna_unos, Xscaled])
```

```python
X_ols=sm.add_constant(Xscaled)
ols = sm.OLS(ytrain, X_ols)
results = ols.fit()
results.summary()
```

```python
Xscaled2=scaler.transform(Xtest)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr=LinearRegression()
lr.fit(Xscaled2,ytest)

y_pred = lr.predict(Xscaled2)
```

```python
r2 = r2_score(ytest, y_pred)
r2
```
