## Regresión polinomial

Observa la siguiente gráfica.

```python
import matplotlib.pyplot as plt; import numpy as np
%matplotlib inline
np.random.seed(0)
x = np.linspace(-3, 3, 100); y = 1 + 4*x**2 + np.random.normal(0, 1, 100)
plt.figure(); plt.xlabel("x"); plt.ylabel("y")
plt.scatter(x, y); plt.grid()
```

Utiliza `sklearn` para realizar una regresión lineal simple ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x $)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X = x.reshape(-1, 1)  
lr=LinearRegression()
lr.fit(X,y)

y_pred = lr.predict(X)

print("Intercepto",lr.intercept_,"Coeficientes",lr.coef_)
```

Crea una gráfica donde se muestren las observaciones como una nube de puntos y la línea que ajustaste con regresión. Calcula el métrico $R^2$.

```python
plt.figure()
plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred, color="red", label="Regresión lineal")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print("R2",r2)
```

Podemos ver que el ajuste no es correcto, pues parece que tiene un comportamiento cuadrático y estamos intentando ajustar una línea recta.

En lugar de hacer una regresión lineal simple, quizás podamos extender el modelo para que tenga varios factores dentro:

$$ Y = \beta_0 + \beta_1 X + \beta_2 X^2 + ... + \beta_p X^p + \epsilon$$

Ajusta un modelo cuadrático con `sklearn` ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x + \hat{\beta_1}x^2 $) y calcula su $R^2$.

```python
polyX2 = PolynomialFeatures(degree=2)
X_2 = polyX2.fit_transform(X)

lr2=LinearRegression()
lr2.fit(X_2,y)

y_pred2 = lr2.predict(X_2)

r2_2 = r2_score(y, y_pred2)

print("Intercepto",lr2.intercept_,"Coeficientes",lr2.coef_,"R2",r2_2)
```

```python
0.619420963180481+4.14388884*(5**2)-0.04955448*(5)
```

Crea una gráfica con las observaciones y la nueva curva que ajustaste.

```python
plt.figure()
plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred2, color="green", label="Regresión polinomial (grado 2)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

Recuerda que una de las metas de ajustar modelos es para poder interpolar y extrapolar. Los datos que tenemos mantienen $x \in [-3, 3]$. ¿Qué pasaría si quisiéramos saber el valor de $y$ cuando $x=5$?

Calcula la predicción $\hat{y}$ para todos los modelos creados (grados 2, 3, 5, 7, 9, y 11).

```python
x2 = np.array([[2]])
x3 = np.array([[3]])
x5 = np.array([[5]])
x7 = np.array([[7]])
x9 = np.array([[9]])
x11 = np.array([[11]])

X2= polyX2.transform(x5)
X3= polyX3.transform(x3)
X5= polyX5.transform(x5)
X7= polyX7.transform(x7)
X9= polyX9.transform(x9)
X11= polyX11.transform(x11)

y_2 = lr2.predict(X2)
y_3 = lr3.predict(X3)
y_5 = lr5.predict(X5)
y_7 = lr7.predict(X7)
y_9 = lr9.predict(X9)
y_11 = lr11.predict(X11)


print("y cuando x=5:", "Grado 2:",y_2, "Grado 3:",y_3,"Grado 5:",y_5,"Grado 7:",y_7,"Grado 9:",y_9,"Grado 11:",y_11)
```

¿Cuál modelo crees que sea mejor?

En este caso, creo que los polinomios de grado 2 o 3 son los más recomendables porque se ajustan bien dentro del rango de $[-3,3]$ y dan valores razonables fuera de él. En cambio, los polinomios de grado alto (7, 9, 11) se disparan fuera del rango y dejan de ser confiables.
